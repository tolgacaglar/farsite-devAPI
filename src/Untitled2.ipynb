```python
import pandas as pd
import geopandas as gpd
```

```python
df = pd.read_pickle('/home/jovyan/data/test_table.pkl')
def change_username_to_jovyan(filepath):
    return filepath[:6] + 'jovyan' + filepath[13:]
df['filepath'] = df['filepath'].apply(change_username_to_jovyan)
```

```python
def geom_from_filepath(filepath):
    if filepath[-3:] == 'shp':
        return gpd.read_file(filepath)['geometry'][0]
    else:
        return None

df['geometry'] = df['filepath'].apply(geom_from_filepath)
df = gpd.GeoDataFrame(df)
```
